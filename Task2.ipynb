```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
# Load image using OpenCV (in BGR format)
image = cv2.imread('/content/rainbow.jpg')  # Replace with your image path
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert to RGB

# Display the original image
plt.imshow(image)
plt.title("Original Image")
plt.axis('off')
plt.show()
```

```python
# Reshape image to (num_pixels, 3)
pixel_data = image.reshape((-1, 3))
pixel_data = np.float32(pixel_data)
```

```python
k = 5  # Try changing this to 2, 3, 8 etc.
kmeans = KMeans(n_clusters=k, random_state=42)
kmeans.fit(pixel_data)
```

```python
# Get cluster centers and labels
centers = np.uint8(kmeans.cluster_centers_)
labels = kmeans.labels_

# Replace each pixel with the center value
segmented_data = centers[labels.flatten()]
segmented_image = segmented_data.reshape(image.shape)

# Show segmented image
plt.imshow(segmented_image)
plt.title(f"Segmented Image with k={k}")
plt.axis('off')
plt.show()
```

```python
# Convert to LAB color space and repeat the steps
lab_image = cv2.cvtColor(image, cv2.COLOR_RGB2LAB)
lab_data = lab_image.reshape((-1, 3))
lab_data = np.float32(lab_data)

kmeans_lab = KMeans(n_clusters=k, random_state=42)
kmeans_lab.fit(lab_data)

lab_centers = np.uint8(kmeans_lab.cluster_centers_)
lab_labels = kmeans_lab.labels_

segmented_lab_data = lab_centers[lab_labels.flatten()]
segmented_lab_image = segmented_lab_data.reshape(image.shape)
segmented_lab_image_rgb = cv2.cvtColor(segmented_lab_image, cv2.COLOR_LAB2RGB)

plt.imshow(segmented_lab_image_rgb)
plt.title(f"LAB Color Segmentation with k={k}")
plt.axis('off')
plt.show()
```
